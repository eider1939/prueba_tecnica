Analizando un poco el DataFrame "Datosventas" encontramos que debemos hacer un proceso de depuracion. 

```python
print(Datosventas.info())
```

```python
#Elimimanos la columna numero uno del dataframe
del Datosventas['Unnamed: 0']
```

```python
#Eliminamos las dos primeras filas del dataframe
Datosventas=Datosventas.drop([0,1],axis=0)
Datosventas=Datosventas.drop([2],axis=0)
```

```python
#renombramos la Columnas que nos quedan
Datosventas=Datosventas.rename(columns={'Unnamed: 1': 'Mes', 'Unnamed: 2': 'Código_Material', 'Unnamed: 3':'Venta_Neta_Kilos','Unnamed: 4':'Venta_Neta_ME'})
```

```python
#cambiamos el typo de datos de las colunas Venta_Neta_Kilos y Venta_Neta_ME ya que estan en objets
# y las necesitamos en int. Tambien comvertimos los valores de esas columnas en valor absoluto ya
# que las ventas deben ser positivas o cero
Datosventas2=Datosventas['Venta_Neta_Kilos'].astype(float).apply(abs).reset_index(name ='Venta_Neta_Kilos')
Datosventas3=Datosventas['Venta_Neta_ME'].astype(float).apply(abs).reset_index(name ='Venta_Neta_ME')
Datosventas2.set_index('index', inplace = True)
Datosventas3.set_index('index', inplace = True)
Datosventas3['Venta_Neta_ME'].apply(abs)
Datosventas.drop(['Venta_Neta_Kilos'], axis=1)
Datosventas=Datosventas.drop(['Venta_Neta_ME'], axis=1)
Datosventas['Venta_Neta_Kilos']=Datosventas2['Venta_Neta_Kilos']
Datosventas['Venta_Neta_ME']=Datosventas3['Venta_Neta_ME']
#se imprime un rango de filas para la verificacion de que ninguna venta sea negativa
Datosventas[3728:3735]
```

```python
print(Datosventas.info())
```

Sea crea un dataframe llamada "Datosventas_1" sel contiene, un dataset agrupado por mes y codigo de material junto con la suma de los kilos del material y el dinero obtenido.

1. Generar un dataset que agrupe por item y mes, desde Enero 2020 a Enero 2022, la cantidad total de ventas en kilos.

```python
Datosventas_Kilos = Datosventas.groupby(['Mes','Código_Material']).Venta_Neta_Kilos.sum().reset_index(name ='Venta_Neta_Kilos')
Datosventas_Kilos
```

```python
#agragamos 1 columnas al archivo xlsx "Fecha" sale de la columna "Mes"
Datosventas['Fecha'] = pd.to_datetime(Datosventas['Mes'], errors='coerce')
#pones a la columna fecha como indice
Datosventas.set_index('Fecha', inplace = True)
Datosventas
```

2. Generar un dataset con el máximo, mínimo y promedio de venta en dinero por cada item desde Enero 2021 a Diciembre 2021.

```python
#Extraer del dataframe las filas que tengan como indice el 2021-01 a 2021-12
#estrae los datos de 2021 de enero a diciembre
Datosventas2021 = Datosventas['2021-01':'2021-12'].iloc[:,0:4]
#generamos 3 dataframe agrupados por mes y codigo del materia y a las ventas se les calcula el min max y mean
Datosventas_min =Datosventas2021.groupby(['Código_Material']).Venta_Neta_ME.min().reset_index(name ='Venta_Neta_ME_Min')
Datosventas_max =Datosventas2021.groupby(['Código_Material']).Venta_Neta_ME.max().reset_index(name ='Venta_Neta_ME_Max')
Datosventas_mean =Datosventas2021.groupby(['Código_Material']).Venta_Neta_ME.mean().reset_index(name ='Venta_Neta_ME_Promedio')
Datosventas_min
```

El dataset Datosventas_min_max nos indicara cual fue la venta minima, la maxima y el promedio, de ventas en el año, por item

```python
#creamos otra data frame donde hacemos un join para unir los data frame de min y max llamado Datosventas_min_max
Datosventas_min_max=pd.merge(Datosventas_min , Datosventas_max,on=['Código_Material'], how='inner')
#se genera otro dataframe en donde a Datosventas_min_max se le añade el daataframe en donde se calculo el promedio
Datosventas_min_max_mean=pd.merge(Datosventas_min_max , Datosventas_mean,on=['Código_Material'], how='inner')
Datosventas_min_max_mean
```
